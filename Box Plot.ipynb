# Box plot

Box plot is pretty much useful to show Quantiles and outliers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(101)
```

```python
salary=np.random.randint(500,10000,495)
gender=np.random.randint(0,2,500)
city=np.random.randint(0,4,500)

# Creating Outliers in salary
salary = np.append(salary,np.random.randint(13000,20000,5))
df= pd.DataFrame(zip(salary,gender,city), columns=['Salary','Gender','City'])
df.head()
```

```python
def gender_mapping(x):
    if(x==1):
        return 'Male'
    else:
        return 'Female'
def city_mapping(x):
    if(x==0): 
        return 'Tokyo'
    elif(x==1): 
        return 'New york'
    elif(x==2): 
        return 'Canada'
df.Gender = df.Gender.map(gender_mapping)
df.City = df.City.map(city_mapping)
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(y='Salary',data=df)
# Or you can write it like this in case you have multiple dataframes
# sns.boxplot(y=df['Salary'])
plt.yscale('linear')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(y='Salary',x='Gender',data=df)
plt.yscale('linear')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(y='Salary',x='Gender',hue='City',data=df)
plt.yscale('linear')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(y='Salary',x='Gender',data=df,whis=1.5)
# whis stands for "Whiskers" which are two limits (upper and lower limits) of the box plot
# They are proportional to the IQR. Whisker value is 1.5 by default (as in the recommended k for outliers definition)
# You can adjust it according to your data and purpose
plt.yscale('linear')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(y='Salary',x='Gender',data=df,whis=2)
plt.yscale('linear')
plt.show()
```

```python
fig = plt.figure(figsize=(4,4))
sns.boxplot(y='Salary',x='Gender',data=df,whis=2, width=0.5)
plt.yscale('linear')
plt.show()
```
