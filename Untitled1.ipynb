```python
import numpy as np
import pandas as pd
np.random.seed(42)
row_1 = np.random.randint(1,100,100)
row_2 = np.random.randint(1,100,100)
row_3 = np.random.randint(1,100,100)
row_4 = np.random.randint(1,100,100)
```

```python
df = pd.DataFrame(zip(row_1,row_2,row_3,row_4), columns=['Class1','Class2','Class3','Class4'])
```

```python
df.head()
```

```python
df.hist(bins=5)
```

```python
plt.hist(df,bins=5,stacked=False)
plt.legend(['Class1','Class2','Class3','Class4'])
plt.ylim(0,35)
plt.xlim(0,100)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Comparison between four classes',loc='center')
plt.show()
```
