# Bar plot

Similar to Histogram but with subtle differences, it exhibits the population of each category in a data.
The differences are:
- Histogram base (x-axis) is continues but Bar chart are categorical (can be numerical, can be nominal)
- Histogram doesn't have space and seperation between ranges but bar charts has spaces between them.

```python
# Prerequisites Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(111)
```

```python
salary = np.random.randint(0,10000+1,500)
gender = np.random.randint(0,2,500)
city= np.random.randint(0,10,500)

df = pd.DataFrame(zip(salary,gender,city),columns=['Salary','Gender','City'])
```

```python
def gender_map(x):
    if(x==0):
        return 'Female'
    elif(x==1):
        return 'Male'
    else:
        return None
def city_map(x):
    match x:
        case 0:
            return 'Syndey'
        case 1:
            return 'Melborne'
        case 2:
            return 'Tehran'
        case 3:
            return 'Isfahan'
        case 4:
            return 'New york'
        case 5:
            return 'Hawaii'
        case 6:
            return 'Canada'
        case 7:
            return 'Toronto'
        case 8:
            return 'Shanghai'
        case 9:
            return 'Beijing'
        case _:
            return None
```

```python
df.Gender = df.Gender.map(gender_map)
df.City = df.City.map(city_map)
```

```python
df.head()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='Gender',hue='Gender',data=df)
plt.title('Data population')
plt.ylabel('Population')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='City',hue='City',data=df)
plt.title('Data population')
plt.ylabel('Population')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='City',hue='Gender',data=df)
plt.title('Data population')
plt.ylabel('Population')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='City',hue='City',palette=sns.color_palette('Paired')[:10],data=df)
plt.title('Data population')
plt.ylabel('Population')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='City',hue='City',palette=sns.color_palette('Set2'),data=df)
plt.title('Data population')
plt.ylabel('Population')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='City',hue='City',palette=sns.color_palette('Paired'),data=df)
plt.title('Data population')
plt.ylabel('Population')
plt.legend(loc='lower right')
plt.show()
```
